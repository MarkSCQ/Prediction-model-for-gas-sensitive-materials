```python
# Get one hot encoding of columns B
data_metal = pd.get_dummies(datacopy.metal, prefix='metal')
data_metal_ox = pd.get_dummies(datacopy.metal_ox, prefix='metal_ox')
data_gas_cate = pd.get_dummies(datacopy.gas_cate, prefix='gas_cate')


# Drop column B as it is now encoded
datacopy = datacopy.drop('metal',axis = 1)
datacopy = datacopy.drop('metal_ox',axis = 1)
datacopy = datacopy.drop('gas_cate',axis = 1)
# Join the encoded df
datacopy = datacopy.join(data_metal)
datacopy = datacopy.join(data_metal_ox)
datacopy = datacopy.join(data_gas_cate)

```
